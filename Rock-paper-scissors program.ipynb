```python
import random

X = 80 # コンピュータの勝率

# この続きを書く
random.seed()

player_win=0
computer_win=0
print('あなたの現在の勝ち点は0です。')
print('私の現在の勝ち点は0です。')

count=1
while count<=10:
    print('第{}試合'.format(count))
    
    while True:
        player=input('あなたの手は？（ぐー＝0,ちょき＝1,ぱー＝2) : ')
        if player=='0' or player=='1' or player=='2':
            player=int(player)
            break
        else:
            print('0,1,2のいずれかを入力してください。')
        
    number_list=[0,1,2]   
    if player==0:
        player_hand='ぐー'
        weight=[(100-X)/2,(100-X)/2,X]
        computer=random.choices(number_list,weights=weight)[0]
    elif player==1:
        player_hand='ちょき'
        weight=[X,(100-X)/2,(100-X)/2]
        computer=random.choices(number_list,weights=weight)[0]
    elif player ==2:
        player_hand='ぱー'
        weight=[(100-X)/2,X,(100-X)/2]
        computer=random.choices(number_list,weights=weight)[0]
        
    if computer==0:
        computer_hand='ぐー'
    elif computer==1:
        computer_hand='ちょき'
    elif computer==2:
        computer_hand='ぱー'
        
    print('あなた:{}'.format(player_hand))
    print('コンピュータ:{}'.format(computer_hand))
    
    
    if computer == player:
        player_win+=1
        computer_win+=1
        print("It's a draw")
        print('あなたの現在の勝ち点は{}です。'.format(player_win))
        print('私の現在の勝ち点は{}です。'.format(computer_win))
        
    elif computer-player==2 or computer-player==-1:
        computer_win+=3
        print('I win!')
        print('あなたの現在の勝ち点は{}です。'.format(player_win))
        print('私の現在の勝ち点は{}です。'.format(computer_win))
        
    elif computer-player==-2 or computer-player==1:
        player_win+=3
        print('You win!')
        print('あなたの現在の勝ち点は{}です。'.format(player_win))
        print('私の現在の勝ち点は{}です。'.format(computer_win)) 
        
    count+=1
    
print('ゲーム終了')
print('あなたの最終的な勝ち点は{}です。'.format(player_win))
print('私の最終的な勝ち点は{}です。'.format(computer_win))
if player_win > computer_win :
    print('You win!')
elif player_win==computer_win :
    print("It's a draw")
elif player_win<computer_win:
    print('I win!')
```
